Este archivo contiene información sobre todos los pasajeros y tripulantes del RMS Titanic. Sus columnas son:

- ``name``: strings with the name of the passenger.
- ``gender``: "male" or "female".
- ``age``: a float with the persons age on the day of the sinking. The age of babies (under 12 months) is given as a fraction of one year.
- ``class``: a string specifying the class (1st, 2nd or 3rd) for passengers or the type of service aboard for crew members.
- ``embarked``: the persons place of of embarkment.
- ``country``: the persons home country.
- ``ticketno``: the persons ticket number (NA for crew members).
- ``fare``: the ticket price (NA for crew members, musicians and employees of the shipyard company).
- ``sibsp``: a number specifying the number if siblings/spouses aboard.
- ``parch``: a number specifying the number of parents/children aboard.
- ``survived``: a string ("no" or "yes") specifying whether the person has survived the sinking.

***Tarifa promedio correspondiente a cada tipo de clase***

````python
#Googleando encontre que se podia usar un groupby para agrupar datos de columnas y asi resumir el código a una linea, con los conocimientos previos habría hecho un código mucho más largo

#Primero agrupo los datos de "class" y después saco la tarifa promedio de cada clase anteriormente agrupada por clase y lo guardo en una variable

promedio_clases = ds_titanic.groupby('class')['fare'].mean()
````

````python
#Acá muestro solo los datos pedidos

print(f'El promedio de tarifa de primera clase es de {promedio_clases[0]:.2f}')
print(f'El promedio de tarifa de segunda clase es de {promedio_clases[1]:.2f}')
print(f'El promedio de tarifa de tercera clase es de {promedio_clases[2]:.2f}')
````

***Encuentre la tasa de supervivencia promedio correspondiente a cada tipo de cabina***

````python
#Acá me fijo cuantos tipos de cabinas hay

ds_titanic["embarked"].unique()
````

````python
#Este ejercicio me complicó bastante, intenté sacar el promedio como hice en el ejercicio anterior y me daba error porque los datos de la columna "survived" no estan en formato int entonces .mean no puede sacar el promedio
#Estuve buscando info y encontre una forma de reemplazar los datos de "survived" por "1" y "0" con un .replace, pero me modificaba los datos de mi DataFrame y si lo volvia a llamar más adelante, volvia con los "1" y 0"
#Finalmente di con la solucion, encontre .map que modifica los elementos de una serie reemplazandolos por los elementos dentro de un diccionario y como trae de vuelta los nuevos valores mapeados, no modifica el original
#Creo una funcion utilizando .map para mapear de nuevo los datos de la columna "survived" y que despues de mapearlos le saque el porcentaje

def supervivientes_promedio(i):
    return i.map({'yes': 1, 'no': 0}).mean()
````

````python
#Acá agrupo por la columna "embarked" y después le aplico la funcion a la columna "survived" utilizando .apply que sirve para aplicar funciones a los DataFrame

supervivientes = ds_titanic.groupby('embarked')['survived'].apply(supervivientes_promedio)
````

````python
#Acá simplemente uso un print para mostrar los resultados finales

print(f'El promedio de supervivencia de B es de {supervivientes[0]:.2f}')
print(f'El promedio de supervivencia de C es de {supervivientes[1]:.2f}')
print(f'El promedio de supervivencia de Q es de {supervivientes[2]:.2f}')
print(f'El promedio de supervivencia de S es de {supervivientes[3]:.2f}')
````
